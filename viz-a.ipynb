```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Load DataFrame

```python
iris = sns.load_dataset("iris")
iris
```

## This data for make a description in detials about iris flower's 

```python
iris.head()
```

## Plot for descripe the relation between Species and sepal length by using violinplot

```python
sns.violinplot(x='species',y='sepal_length',data=iris)
```

## boxplot to descripe all attribute's for given species

```python
sns.boxplot(data=iris)
```

## descripe flowers as pairplot

```python
sns.pairplot(iris, hue="species")
```

## Conclusion
- Setosa species it's the smallest compare to the other's
- petal length have the higher rank between all species to be common
- as violinplot the virginica flower have the longest leaf
