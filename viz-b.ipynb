```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Load DataFrame

```python
tips = sns.load_dataset("tips")
tips

```

## This data for descripe which of the two gender smoke more and how many bill's they pay for , also it descripe the spetial time for smoking more ,...etc

```python
tips.head()
```

## making scattor plot for the ralation between tip's and total bill's , by using lmplot 

```python
sns.lmplot(x='tip' ,y='total_bill',data=tips)
```

## Descripe the Smoker people as boxplot Which show bills by day and time , also male and female below

```python
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
```

```python
sns.boxplot(x="sex", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
```
